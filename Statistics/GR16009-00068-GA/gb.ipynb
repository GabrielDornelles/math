```python
import numpy as np
import csv
import collections
import matplotlib.pyplot as plt
from collections import Counter
from matplotlib.pyplot import figure
from methods import build_dict_from_csv, check_both, check_occurence, histograms
import seaborn as sns
# Custo de MÃ£o de Obra (X1), Custo de Entrega (X2), Total de
#Desconto (X3), Valor do Custo (X4) e Valor de Venda (X5).
keys = ["CustoMaoDeObra", "CustoEntrega", "TotalDesconto", "ValorCusto", "ValorVenda"]
```

```python
general_dict = build_dict_from_csv()
Y = np.array([float(x) for x in general_dict["Lucro"]])
color_dict = dict({'CustoMaoDeObra':'brown',
                  'CustoEntrega':'green',
                  'TotalDesconto': 'orange',
                  'ValorCusto': 'red',
                   'ValorVenda': 'dodgerblue'})
```

```python
for x in keys:
    X = np.array([float(x) for x in general_dict[x]])
    m, b = np.polyfit(X, Y, 1)
    figure(dpi=150)
    plt.style.use('default')
    plt.xlabel(x)
    plt.ylabel("lucro")
    plt.scatter(X,Y,color=color_dict[x],s=1)
    plt.plot(X, m*X + b,color="dimgray",linewidth=0.5)
    plt.legend(["best fit","points"])
    plt.show()
```
